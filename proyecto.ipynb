```python
df['gama'] = np.where(df['price'] > 30000, 1, 0)
```

```python
columns = ['symboling','fueltype', 'aspiration',
'doornumber', 'carlength', 'carwidth', 
'carheight', 'curbweight', 'enginetype', 
'cylindernumber', 'enginesize', 'fuelsystem', 
'horsepower', 'peakrpm', 'citympg', 'highwaympg','gama']
```

```python
df = df[columns]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filtrar el DataFrame para obtener subconjuntos de gama alta y baja
df_alta_gama = df[df['gama'] == True]
df_baja_gama = df[df['gama'] == False]

# FunciÃ³n para graficar caracterÃ­sticas para gama alta y baja
def plot_feature_distribution(feature):
    plt.figure(figsize=(12, 6))
    
    # Histograma para gama alta
    plt.subplot(1, 2, 1)
    if feature in ['carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'horsepower', 'peakrpm']:
        sns.histplot(df_alta_gama[feature], bins=5, color='skyblue', kde=True)
    else:
        sns.countplot(x=feature, data=df_alta_gama, palette='Set1')
    plt.title(f'DistribuciÃ³n en Gama Alta - {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frecuencia')

    # Histograma para gama baja
    plt.subplot(1, 2, 2)
    if feature in ['carlength', 'carwidth', 'carheight', 'curbweight', 'enginesize', 'horsepower', 'peakrpm']:
        sns.histplot(df_baja_gama[feature], bins=5, color='lightcoral', kde=True)
    else:
        sns.countplot(x=feature, data=df_baja_gama, palette='Set2')
    plt.title(f'DistribuciÃ³n en Gama Baja - {feature}')
    plt.xlabel(feature)
    plt.ylabel('Frecuencia')

    plt.tight_layout()
    plt.show()

# Iterar sobre las columnas y graficar la distribuciÃ³n para cada una
for column in columns[:-1]:  # Excluyendo la columna 'gama'
    plot_feature_distribution(column)
```

```python
alta = df.query("gama==1")
baja = df.query("gama==0")
for column in df.columns:
    try:
        print(f"Columna: {column}")
        print(f"ALTA: {alta[column].value_counts()}")
        print(f"BAJA: {baja[column].value_counts()}")
        print()
    except:
        print(f"Columna {column} es de tipo numÃ©rico")
        print(f"ALTA: {alta[column].describe}")
        print(f"BAJA: {baja[column].describe}")
```
